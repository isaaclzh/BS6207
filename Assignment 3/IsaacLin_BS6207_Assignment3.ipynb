## Import Libraries

```python
import torch
import random
from collections import OrderedDict
import torchvision
import numpy as np
import math
```

## Generate Randomized Inputs

```python
# Generate inputs
torch.manual_seed(4)
pixel_input = torch.randn(1,3,32,32)
pixel_input
```

## Model 1

```python
m1 = torch.nn.MaxPool2d(kernel_size=2, stride=1, padding=0,
                        dilation=1, return_indices=False, ceil_mode=False)
torch_out = m1(pixel_input)
torch_out
```

```python
def model1(x):
    result = []
    for k in range(3): 
        img = np.zeros((31,31))
        for i in range(31):
            for j in range(31):
                temp = np.max(x[0][k][i:i+2,j:j+2])
                img[i][j] = temp
        result.append(img)
    return result
```

```python
my_out = model1(np.array(pixel_input))
my_out
```

## Model 2

```python
m2 = torch.nn.AvgPool2d(kernel_size=2, stride=1, padding=0, ceil_mode=False, 
                        count_include_pad=True, divisor_override=None)
torch_out = m2(pixel_input)
torch_out
```

```python
def model2(x):
    result = []
    for k in range(3): 
        img = np.zeros((31,31))
        for i in range(31):
            for j in range(31):
                temp = np.mean(x[0][k][i:i+2,j:j+2])
                img[i][j] = temp
        result.append(img)
    return result
```

```python
my_out = model2(np.array(pixel_input))
my_out
```

## Model 3

```python
m3 = torch.nn.Conv2d(in_channels=3, out_channels=6,kernel_size=3, stride=1, 
                     padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros')

m3.weight.data.fill_(0.01)
m3.bias.data.fill_(0.02)
torch_out = m3(pixel_input)
torch_out
```

```python
def model3(x,w,b):
    result = []
    for k in range(6): 
        img = np.zeros((30,30))
        for i in range(30):
            for j in range(30):
                temp = []
                for g in range(3):
                    temp.append(x[0][g][i:i+3,j:j+3] * w)
                    img[i][j] = np.sum(temp) + b
        result.append(img)
    return result
```

```python
my_out = model3(np.array(pixel_input),0.01,0.02)
my_out
```

## Model 4

```python
m4 = torch.nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5, stride=2, 
                     padding=0, dilation=2, groups=1, bias=True, padding_mode='zeros')

m4.weight.data.fill_(0.01)
m4.bias.data.fill_(0.02)
torch_out = m4(pixel_input)
torch_out
```

```python
def model4(x,w,b):
    result = []
    for k in range(6): 
        img = np.zeros((12,12))
        for i in range(0,24,2):
            for j in range(0,24,2):
                temp = []
                for g in range(3):
                    temp.append(x[0][g][i:i+9,j:j+9] * w)
                    img[i//2][j//2] = np.sum(temp) + b
        result.append(img)
    return result
```

```python
w = np.zeros((9,9))
for i in range(0,9,2):
    for j in range(0,9,2):
        w[i][j] = 0.01
my_out = model4(np.array(pixel_input),w,0.02)
my_out
```

## Model 5

```python
m5 = torch.nn.ConvTranspose2d(in_channels=3, out_channels=4, kernel_size=3, stride=1, padding=0, 
                              output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros')

m5.weight.data.fill_(0.01)
m5.bias.data.fill_(0.02)
torch_out = m5(pixel_input)
torch_out
```

```python
def model5(x,w,b):
    result = []
    for k in range(4): 
        img = np.zeros((34,34))
        for i in range(34):
            for j in range(34):
                temp = []
                for g in range(3):
                    temp.append(x[g][i:i+3,j:j+3] * w)
                    img[i][j] = np.sum(temp) + b
        result.append(img)
    return result
```

```python
pixel_inputT = []
for k in range(3):
    tempT= np.zeros((36,36))  
    for i in range(2,34):
        for j in range(2,34):
            tempT[i][j] = np.array(pixel_input)[0][k][i - 2][j - 2]
    pixel_inputT.append(tempT)
my_out = model5(np.array(pixel_inputT),0.01,0.02)
my_out
```

## Model 6

```python
def model6(x):
    result = []
    for k in range(3):
        for i in range(32):
            for j in range(32):
                result.append(x[0][k][i][j])
    return result
my_out = model6(np.array(pixel_input))
my_out
```

## Model 7

```python
torch_out = torch.sigmoid(pixel_input,out=None)
torch_out
```

```python
def model7(x):
    def sigmoid(x):
        return 1 / (1 + math. exp(-x))
    result = []
    for k in range(3):
        temp = np.zeros((32,32))
        for i in range(32):
            for j in range(32):
                temp[i][j] = sigmoid(x[0][k][i][j])
        result.append(temp)
    return result
my_out = model7(np.array(pixel_input))
my_out
```

## Model 9

```python
running_mean = torch.zeros(3) + 1
running_var = torch.ones(3) * 2
torch_out = torch.nn.functional.batch_norm(pixel_input, running_mean, running_var, weight=None, 
                                           bias=None, training=False, momentum=0.1, eps=1e-05)
torch_out
```

```python
def model9(x,mean,var,eps):
    result = []
    for k in range(3):
        img = np.zeros((32,32))
        for i in range(32):
            for j in range(32):
                img[i][j] = (x[0][k][i][j] - mean)/ np.sqrt(var + eps)
        result.append(img)
    return result
```

```python
my_out = model9(np.array(pixel_input),1,2,1e-05)
my_out
```

## Model 10

```python
# Generate inputs
torch.manual_seed(4)
pixel_input = torch.FloatTensor(1,32).uniform_(0,1)
target = torch.randint(32, (1,))

torch_out = torch.nn.functional.cross_entropy(pixel_input, target, weight=None, size_average=None, 
                                              ignore_index=-100, reduce=None, reduction='mean')
torch_out
```

```python
def model10(x,target):
    temp = 0
    for i in range(32):
        temp += np.exp(x[0][i])    
    result = -1 * x[0][target[0]] + np.log(temp)
    return result
```

```python
my_out = model10(np.array(pixel_input),np.array(target))
round(my_out,4)
```

## Model 11

```python
# Generate inputs
torch.manual_seed(4)
pixel_input = torch.randn(1,3,32,32)
target = torch.randn(1,3,32,32)

torch_out = torch.nn.functional.mse_loss(pixel_input, target, size_average=None, 
                                   reduce=None, reduction='mean')
torch_out
```

```python
def model11(x,target):
    result = []
    for k in range(3):
        for i in range(32):
              for j in range(32):
                    result.append(np.square(x[0][k][i][j] - target[0][k][i][j]))
    return np.mean(result)
```

```python
my_out = model11(np.array(pixel_input),np.array(target))
round(my_out,4)
```
