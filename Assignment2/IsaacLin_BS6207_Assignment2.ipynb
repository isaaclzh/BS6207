```python
import numpy as np
```

```python
import math

alpha = 0.3
beta1 = 0.9
beta2 = 0.999
epsilon = 0
m = 0 
v = 0 
res = 0
```

```python
def adam(t, dw, m_dw, v_dw):
    # Update the first moment estimate
    m_dw = beta1 * m_dw + (1 - beta1) * dw
    # Update the second moment estimate
    v_dw = beta2 * v_dw + (1 - beta2) * (dw ** 2)
    
    # Bias correction
    m_dw_corr = m_dw / (1 - beta1 ** t)
    v_dw_corr = v_dw / (1 - beta2 ** t)
    res = (alpha * (m_dw_corr / (np.sqrt(v_dw_corr) + epsilon)))
    
    # Update weights and bias
    return m_dw, v_dw, res
```

```python
# h = 1
for i in range(1,100):
    if res < 1:
        dw = -1
    elif res > 1 and res - 1 < 1:
        dw = 1
    elif res - 1 > 1:
        dw = -1
    
    m, v, sita = adam(i,dw,m,v)
    res = res-sita
    print(round(res,2))
```
